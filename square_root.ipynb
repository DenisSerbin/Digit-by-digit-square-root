```python
from copy import deepcopy

class LongNumber:
    
    def __init__(self, num = None):
        if not num or num == 0:
            self.digits = []
        else:
            self.digits = [int(x) for x in str(num)]

    def __repr__(self):
        if not self.digits:
            return "0"
        else:
            return ''.join([str(x) for x in self.digits])
    
    def __add__(self, digit):
        
        new_num = LongNumber()
        new_num.digits = deepcopy(self.digits)
        
        if (type(digit) is not int) or digit >= 10 or digit <= 0:
            return new_num
        
        if not new_num.digits:
            new_num.digits.append(digit)
            return new_num
        else:
            transfer = 0
            for i in range(len(new_num.digits)):
                if i: digit = 0
                if new_num.digits[-1 - i] + digit + transfer <= 9:
                    new_num.digits[-1 - i] += digit + transfer
                    transfer = 0
                    break
                else:
                    new_num.digits[-1 - i] = (new_num.digits[-1 - i] + digit + transfer) % 10
                    transfer = 1
            if transfer:
                new_num.digits.insert(0, transfer)
            return new_num
    
    def __mul__(self, other):
        
        new_num = LongNumber()
        new_num.digits = deepcopy(self.digits)
        
        if (type(other) is not int) or other > 10 or other < 0:
            return new_num
        
        if other == 0:
            return LongNumber()
        
        if other == 10:
            new_num.digits.append(0)
            return new_num
        
        if not new_num.digits:
            return LongNumber()
        
        transfer = 0
        for i in range(len(new_num.digits)):
            prod = new_num.digits[-1 - i] * other + transfer
            new_num.digits[-1 - i] = prod % 10
            transfer = int(prod / 10)
        
        if transfer:
            new_num.digits.insert(0, transfer)
        
        return new_num
        
    def __sub__(self, other):
        
        new_num = LongNumber()
        new_num.digits = deepcopy(self.digits)
        
        if not isinstance(other, LongNumber):
            print("Wrong type")
            return new_num
        
        #if not self >= other:
        #    print("Not less than")
        #    return new_num
        
        if not self.digits or not other.digits:
            return new_num
        
        transfer = 0
        for i in range(len(other.digits)):
            diff = new_num.digits[-1 - i] + transfer - other.digits[-1 - i]
            if diff >= 0:
                new_num.digits[-1 - i] = diff
                transfer = 0
            else:
                new_num.digits[-1 - i] = diff + 10
                transfer = -1
        
        while transfer < 0 and i < len(new_num.digits):
            diff = new_num.digits[-1 - i - 1] + transfer
            if diff >= 0:
                new_num.digits[-1 - i - 1] = diff
                transfer = 0
            else:
                new_num.digits[-1 - i - 1] = diff + 10
                transfer = -1
            i += 1
        
        while new_num.digits and new_num.digits[0] == 0:
            new_num.digits.pop(0)
            
        return new_num
    
    def __le__(self, other):
        
        if not isinstance(other, LongNumber):
            print("Wrong type")
            return False
        
        if len(self.digits) < len(other.digits):
            return True
        
        if len(self.digits) > len(other.digits):
            return False
        
        for i in range(len(other.digits)):
            if self.digits[i] < other.digits[i]:
                return True
            elif self.digits[i] > other.digits[i]:
                return False
            else:
                pass
        
        return True
    
    def length(self):
        return len(self.digits)
```

```python
from IPython.core.debugger import set_trace
import math

def DigitByDigitSqRt(num, precision = None):
    
    if num == 0:
        print("0")
        return
    
    num_str = str(num)
    digits_processed = 1 if len(num_str) % 2 else 2
    c_int = int(num_str[0]) if len(num_str) % 2 else int(num_str[ : 2])
    c = LongNumber(c_int)
    p = LongNumber()
    
    dot = False
    max_step = precision if precision else math.inf
    step = 0
    
    while step <= max_step:
        left = 0
        right = 9
        x = 0
        while left <= right:
            mid = (right + left) // 2
            if ((p*2)*10 + mid)*mid <= c:
                left = mid + 1
                x = mid
            else:
                right = mid - 1
        
        y = ((p*2)*10 + x)*x
        p = p*10 + x
        c = c - y
        
        #set_trace()
        if c.length() == 0 and digits_processed == len(num_str):
            print(x, end = '', flush = True)
            return
        elif c.length() == 0 and digits_processed < len(num_str):
            print(x, end = '', flush = True)
            if num_str[digits_processed] != '0':
                c = LongNumber(int(num_str[digits_processed : digits_processed + 2]))
            else:
                c = LongNumber(int(num_str[digits_processed + 1]))
            digits_processed += 2
        elif c.length() and digits_processed < len(num_str):
            print(x, end = '', flush = True)
            c = (c*10) + int(num_str[digits_processed])
            c = (c*10) + int(num_str[digits_processed + 1])
            digits_processed += 2
        elif c.length() and digits_processed == len(num_str):
            print(x, end = '', flush = True)
            if not dot:
                print(".", end = '', flush = True)
                dot  = True
            c = (c*10)*10
            step += 1
    
    return
```
